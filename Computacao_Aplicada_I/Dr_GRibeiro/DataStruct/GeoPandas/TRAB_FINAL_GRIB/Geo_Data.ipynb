```python
import geopandas as gpd
import matplotlib.pyplot as plt

amazonia_legal = gpd.read_file(r'data/amazonia_legal/amazlegal.shp')
para_municipios = gpd.read_file(r'data/pa_municipios/15MUE250GC_SIR.shp')
poly_desforestation = gpd.read_file(r'data/Deter_201711_shp/Deter_201711.shp')

fig, ax = plt.subplots(figsize = (25,15)) 
amazonia_legal.plot(cmap = 'Purples',ax=ax, legend=True)
amazonia_legal.geometry.boundary.plot(color=None,edgecolor='k',linewidth = 1,ax=ax)
para_municipios.plot(ax=ax, color='#CCFFFF', edgecolor='black')
para_municipios.geometry.boundary.plot(color='black',edgecolor='k',linewidth = 1,ax=ax)
poly_desforestation.plot(cmap = 'Reds',ax=ax)
poly_desforestation.geometry.boundary.plot(color=None,edgecolor='k',linewidth = 5,ax=ax)
```

```python
points = poly_desforestation.geometry
xx = [x.centroid.coords.xy[1][0] for x in points.boundary]
yy = [x.centroid.coords.xy[0][0] for x in points.boundary]
coords = ''
for i in range(len(xx)):
    coords += str(xx[i]) + ',' + str(yy[i]) + '\n'
with open('points.txt','a') as f:
    f.write(coords)
```
