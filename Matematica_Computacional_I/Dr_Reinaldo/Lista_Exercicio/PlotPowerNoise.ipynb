```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import thinkplot
from scipy.interpolate import spline
```

```python
whiteNormalize = pd.read_csv('dados/WhiteNormalize.csv')
whiteRandpower = pd.read_csv('dados/WhiteRandpower.csv')

pinkNormalize = pd.read_csv('dados/with_1_2/PinkNormalize.csv')
pinkRandpower = pd.read_csv('dados/with_1_2/PinkRandpower.csv')

redNormalize = pd.read_csv('dados/with_1_2/RedNormalize.csv')
redRandpower = pd.read_csv('dados/with_1_2/RedRandpower.csv')
```

```python
plt.figure(figsize=(20,5))
plt.plot(whiteNormalize, linewidth=0.5)
plt.title('Normalized WhiteNoise')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(whiteRandpower, linewidth=0.5)
plt.title('Rand WhiteNoise')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(whiteRandpower, linewidth=0.6)
plt.title('Randpower WhiteNoise')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(pinkNormalize, linewidth=0.6)
plt.title('Normalize PinkNoise')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(pinkRandpower, linewidth=0.6)
plt.title('Rand PinkNoise')
plt.show()

fig = pinkRandpower.rolling(window=50, win_type='gaussian', center=True).mean(std=100).plot(label='Red Noise', figsize = (20,5))
```

```python
plt.figure(figsize=(20,5))
plt.plot(redNormalize, linewidth=2)
plt.title('Normalize RedNoise')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(redRandpower, linewidth=2)
plt.title('Rand RedNoise')
plt.show()
```
